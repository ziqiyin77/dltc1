```python
import torch
from torch.utils.data import DataLoader, random_split
from torchvision import transforms
from torchvision.datasets import ImageFolder

# Paths
IMAGE_DIR = "Data/images_original"

# Hyperparameters
BATCH_SIZE = 32
SEED = 42

# Image preprocessing required by the coursework
image_transform = transforms.Compose([
    transforms.Resize((180, 180)),
    transforms.ToTensor()
])

# Load spectrogram image dataset
image_dataset = ImageFolder(
    root=IMAGE_DIR,
    transform=image_transform
)

# Check classes
print("Classes:", image_dataset.classes)
print("Class to index:", image_dataset.class_to_idx)
print("Total image samples:", len(image_dataset))

# Train / validation / test split: 70% / 20% / 10%
total_size = len(image_dataset)
train_size = int(0.7 * total_size)
val_size = int(0.2 * total_size)
test_size = total_size - train_size - val_size

train_dataset, val_dataset, test_dataset = random_split(
    image_dataset,
    [train_size, val_size, test_size],
    generator=torch.Generator().manual_seed(SEED)
)

print("Train size:", len(train_dataset))
print("Validation size:", len(val_dataset))
print("Test size:", len(test_dataset))

# DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False
)

# Check one batch
images, labels = next(iter(train_loader))

print("Image batch shape:", images.shape)
print("Label batch shape:", labels.shape)
print("Example labels:", labels[:10])
```

```python
# ============================================================
# Image models for coursework (Net1 to Net4)
# This section uses the existing train_loader, val_loader, test_loader
# created in the preprocessing cell above.
# ============================================================

import copy
import pandas as pd
import torch.nn as nn
import torch.optim as optim

# Reproducibility and device setup
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
NUM_CLASSES = 10
INPUT_SHAPE = (3, 180, 180)
LEARNING_RATE = 1e-3
print("Using device:", DEVICE)
```

```python
# ------------------------------------------------------------
# Reusable helper functions
# ------------------------------------------------------------

def evaluate_model(model, data_loader, criterion=None, device=DEVICE):
    """
    Evaluate a model on a dataloader.

    Returns:
        avg_loss (float or None): Average loss over the loader if criterion is provided.
        accuracy (float): Classification accuracy in percentage.
    """
    model.eval()
    total_correct = 0
    total_samples = 0
    total_loss = 0.0

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)

            if criterion is not None:
                loss = criterion(outputs, labels)
                total_loss += loss.item() * images.size(0)

            predictions = outputs.argmax(dim=1)
            total_correct += (predictions == labels).sum().item()
            total_samples += labels.size(0)

    avg_loss = (total_loss / total_samples) if criterion is not None else None
    accuracy = 100.0 * total_correct / total_samples
    return avg_loss, accuracy


def train_model(model, train_loader, val_loader, epochs, optimizer, criterion, device=DEVICE):
    """
    Train a model and track metrics for each epoch.

    Tracks:
    - training loss
    - validation loss
    - validation accuracy

    Returns:
        best_model_state (dict): Model weights with best validation accuracy.
        history (dict): Metric history for plotting/reporting.
    """
    history = {
        "train_loss": [],
        "val_loss": [],
        "val_accuracy": []
    }

    best_val_acc = -1.0
    best_model_state = copy.deepcopy(model.state_dict())

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        total_samples = 0

        for images, labels in train_loader:
            images = images.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)
            total_samples += images.size(0)

        train_loss = running_loss / total_samples
        val_loss, val_acc = evaluate_model(model, val_loader, criterion=criterion, device=device)

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["val_accuracy"].append(val_acc)

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_state = copy.deepcopy(model.state_dict())

        print(
            f"Epoch [{epoch + 1:3d}/{epochs}] "
            f"Train Loss: {train_loss:.4f} | "
            f"Val Loss: {val_loss:.4f} | "
            f"Val Acc: {val_acc:.2f}%"
        )

    return best_model_state, history
```

```python
# ------------------------------------------------------------
# Model definitions
# ------------------------------------------------------------

class Net1FC(nn.Module):
    """Net1: fully connected network with two hidden layers."""
    def __init__(self, num_classes=NUM_CLASSES):
        super().__init__()
        self.flatten = nn.Flatten()
        self.classifier = nn.Sequential(
            nn.Linear(3 * 180 * 180, 512),
            nn.ReLU(),
            nn.Linear(512, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.flatten(x)
        return self.classifier(x)


class Net2CNN(nn.Module):
    """
    Net2: CNN baseline following a simple Figure-1 style stack:
    Conv -> ReLU -> MaxPool -> Conv -> ReLU -> MaxPool -> FC layers.
    """
    def __init__(self, num_classes=NUM_CLASSES):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        # Infer flattened size from the known input shape
        with torch.no_grad():
            dummy = torch.zeros(1, *INPUT_SHAPE)
            feature_dim = self.features(dummy).view(1, -1).size(1)

        self.classifier = nn.Sequential(
            nn.Linear(feature_dim, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)


class Net3CNNBatchNorm(nn.Module):
    """Net3: same CNN as Net2, but with Batch Normalisation layers."""
    def __init__(self, num_classes=NUM_CLASSES):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        with torch.no_grad():
            dummy = torch.zeros(1, *INPUT_SHAPE)
            feature_dim = self.features(dummy).view(1, -1).size(1)

        self.classifier = nn.Sequential(
            nn.Linear(feature_dim, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)


# Net4 uses the same architecture as Net3 and differs in optimizer choice.
class Net4CNNBatchNormRMSProp(Net3CNNBatchNorm):
    """Net4: same architecture as Net3; to be trained with RMSProp."""
    pass
```

```python
# ------------------------------------------------------------
# Training runner for Net1-Net4 at 50 and 100 epochs
# ------------------------------------------------------------

criterion = nn.CrossEntropyLoss()

# Factories to create fresh models and optimizers for each run
model_factories = {
    "Net1": lambda: Net1FC(num_classes=NUM_CLASSES),
    "Net2": lambda: Net2CNN(num_classes=NUM_CLASSES),
    "Net3": lambda: Net3CNNBatchNorm(num_classes=NUM_CLASSES),
    "Net4": lambda: Net4CNNBatchNormRMSProp(num_classes=NUM_CLASSES),
}


def build_optimizer(model_name, model):
    # Net4 must use RMSProp; others use Adam for a stable default
    if model_name == "Net4":
        return optim.RMSprop(model.parameters(), lr=LEARNING_RATE)
    return optim.Adam(model.parameters(), lr=LEARNING_RATE)


epoch_options = [50, 100]
results = []
all_histories = {}

for model_name, model_fn in model_factories.items():
    for epochs in epoch_options:
        print()
        print("=" * 70)
        print(f"Training {model_name} for {epochs} epochs")
        print("=" * 70)

        model = model_fn().to(DEVICE)
        optimizer = build_optimizer(model_name, model)

        best_state, history = train_model(
            model=model,
            train_loader=train_loader,
            val_loader=val_loader,
            epochs=epochs,
            optimizer=optimizer,
            criterion=criterion,
            device=DEVICE,
        )

        # Evaluate the best validation checkpoint on validation and test sets
        model.load_state_dict(best_state)
        final_val_loss, final_val_acc = evaluate_model(model, val_loader, criterion=criterion, device=DEVICE)
        _, test_acc = evaluate_model(model, test_loader, criterion=criterion, device=DEVICE)

        run_key = f"{model_name}_{epochs}epochs"
        all_histories[run_key] = history

        results.append({
            "model": model_name,
            "epochs": epochs,
            "optimizer": optimizer.__class__.__name__,
            "final_train_loss": history["train_loss"][-1],
            "final_val_loss": final_val_loss,
            "final_val_accuracy": final_val_acc,
            "test_accuracy": test_acc,
        })

        print(
            f"Completed {run_key} | "
            f"Val Loss: {final_val_loss:.4f} | "
            f"Val Acc: {final_val_acc:.2f}% | "
            f"Test Acc: {test_acc:.2f}%"
        )
```

```python
# ------------------------------------------------------------
# Save final results for coursework report
# ------------------------------------------------------------

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by=["model", "epochs"]).reset_index(drop=True)

print(results_df)

# Save required CSV output
results_df.to_csv("results_image_models.csv", index=False)
print("Saved results to results_image_models.csv")
```
